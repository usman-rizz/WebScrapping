# **Data Preprocessing**

```python
import pandas as pd
import numpy as np
import random
```

```python
# loading the dataset
data = pd.read_csv('ExtractedData.csv')
```

```python
# Making the copy of dataset to seperate cleaned and raw data
cleaned_data = data.copy()
```

## **Exploring Raw Dataset**

```python
cleaned_data["Reviews"] = cleaned_data["Reviews"].str.extract(r"(\d+)")
```

```python
cleaned_data.sample(3)
```

```python
# Fill NaN in Rating with random number between 1 and 5
cleaned_data["Rating"] = cleaned_data["Rating"].fillna("").apply(lambda x: x if x != "" else round(random.uniform(1, 5), 1))
```

```python
cleaned_data.sample(6)
```

```python
# checkig for duplicates
print(cleaned_data.duplicated().sum())

# removing duplicates
cleaned_data = cleaned_data.drop_duplicates()
```

```python
cleaned_data.to_csv('CleanedData.csv', index=False)
```
